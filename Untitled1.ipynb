# Modelo

```python
class RNAModel:
    def __init__(self, dados_referencia: pd.DataFrame, classes_referencia: pd.DataFrame, hidden_neurons: int = 500, output_neurons: int = 9):
        self.dados_referencia = dados_referencia
        self.classes_referencia = classes_referencia
        self.hidden_neurons = hidden_neurons
        self.output_neurons = output_neurons
        self.input_neurons = self.dados_referencia.shape[1]

    # Inicializando pesos e bias
        self.W1 = np.random.randn(self.input_neurons, self.hidden_neurons) / np.sqrt(self.input_neurons)
        self.B1 = np.zeros((1, self.hidden_neurons))
        self.W2 = np.random.randn(self.hidden_neurons, self.output_neurons) / np.sqrt(self.hidden_neurons)
        self.B2 = np.zeros((1, self.output_neurons))
        self.model_dict = {'W1': self.W1, 'B1': self.B1, 'W2': self.W2, 'B2': self.B2}
        self.z1 = 0
        self.f1 = 0
    
    def forward(self, dados_referencia: pd.DataFrame) -> np.ndarray:
        # Equação da reta
        self.z1 = dados_referencia.dot(self.W1) + self.B1

        # Função de Ativação
        self.f1 = np.tanh(self.z1)

        # Equação da reta 2
        z2 = self.f1.dot(self.W2) + self.B2

        # Softmax
        z2_max = np.max(z2, axis=1, keepdims=True)  # Subtraindo o máximo de cada linha
        exp_values = np.exp(z2 - z2_max)  # Subtração evita overflow
        softmax = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        
        return softmax


    def backpropagation(self, softmax: np.ndarray, learning_rate: float) -> None:
        delta2 = np.copy(softmax)
        delta2[range(self.dados_referencia.shape[0]), self.classes_referencia] -= 1
        dW2 = (self.f1.T).dot(delta2)
        dB2 = np.sum(delta2, axis = 0, keepdims=True)
        delta1 = delta2.dot(self.W2.T)*(1-np.power(np.tanh(self.z1),2))
        dW1 = (self.dados_referencia.T).dot(delta1)
        dB1 = np.sum(delta1, axis = 0, keepdims=True)

    # Atualização pesos e bias

        self.W1 += - learning_rate*dW1
        self.W2 += - learning_rate*dW2
        self.B1 += - learning_rate*dB1
        self.B2 += - learning_rate*dB2
        
    def fit(self, epochs: int, lr: float):

        for epoch in range(epochs):

            outputs = self.forward(self.dados_referencia)
            self.backpropagation(outputs, lr)

            # Acurácia
            prediction = np.argmax(outputs, axis=1)
            correct_classes = self.classes_referencia.flatten()  # Garantir que seja 1D
            correct = (prediction == correct_classes).sum()  # Comparação elemento a elemento
            accuracy = correct / self.classes_referencia.shape[0]

            print(f'Epoch: [{epoch + 1} / {epochs}] Accuracy: {accuracy:.5f}')

        return prediction
```

```python
modelo = RNAModel(dados_referencia, classes_referencia, hidden_neurons=500, output_neurons=9)
modelo.fit(epochs=30, lr=0.0001)

predicoes = modelo.forward(dados_novos).argmax(axis=1)
```
